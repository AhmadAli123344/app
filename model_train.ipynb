```python
import pandas as pd
df=pd.read_csv('heart (1).csv')

from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MinMaxScaler,LabelEncoder,OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import pickle
df.info()
```

```python
# df['ST_Slope'].value_counts()
```

```python
X=df.drop(['HeartDisease','ExerciseAngina'],axis=1)
X
y=df['HeartDisease']

X_numeric=['Age','RestingBP','Cholesterol','FastingBS','MaxHR','Oldpeak']
X_categoric=['Sex','ChestPainType','RestingECG','ST_Slope']

a=Pipeline(steps=[('num',MinMaxScaler())])
b=Pipeline(steps=[('cat',OneHotEncoder())])

preprocess=ColumnTransformer(transformers=[('apply1',a,X_numeric),('apply2',b,X_categoric)])
```

```python
a=X.iloc[1]
import numpy as np
b=np.array(a)
b.reshape(1,-1)
```

```python
model=Pipeline(steps=[('preprocess',preprocess),('predict',LogisticRegression())])
model.fit(X,y)
with open('model_heartdisease.pkl', 'wb') as file:
    pickle.dump(model, file)
```
